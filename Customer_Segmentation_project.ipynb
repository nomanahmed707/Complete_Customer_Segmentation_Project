#Implementing K_mean Clustering on the Dataset

```python
# Implementing K_mean Clustering on the Dataset
from sklearn.preprocessing import LabelEncoder
label_en = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = label_en.fit_transform(df[col])
```

```python
df.head(5)
```

```python
from sklearn.cluster import KMeans
features = df[['Age',
               'Purchase Amount (USD)',
               'Review Rating',
               'Previous Purchases',
               'Frequency of Purchases', "Item Purchased"]]
```

```python
features
```

# Normalizing the Features

```python
# Normalizing the features
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
features = ss.fit_transform(features)
```

```python
features.shape
```

```python
# Find optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the graph
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fit KMeans with chosen cluster number (say 8)
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
df['Cluster']=kmeans.fit_predict(features)
```

```python
df['Cluster'].shape
```

```python
# Visualize clusters
plt.scatter(df['Age'], df['Purchase Amount (USD)'], c=df['Cluster'], cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.title('Clusters')
plt.show()
```

```python
# Visualize clusters 3D
import plotly.express as px
fig = px.scatter_3d(df, x='Age', y='Purchase Amount (USD)', z='Review Rating', color='Cluster', title="Clusters Visualization in 3D")
fig.show()

```

# Clusters Evaluation

```python
df['Cluster']
```

```python
# Evaluating the Predicted Clusters
from sklearn.metrics import silhouette_score
silhouette_score(features, df['Cluster'])
```

# Using DBSCAN

```python
features
```

```python
# Using DBSCAN
from sklearn.cluster import DBSCAN
# Fit DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=7)  # You can tune these values
labels = dbscan.fit_predict(features)

# Add labels to DataFrame
df['DBSCAN_Cluster'] = labels
```

```python
# Remove noise points (-1) for Silhouette evaluation
filtered_features = features[labels != -1]
filtered_labels = labels[labels != -1]

if len(set(filtered_labels)) > 1:
    score = silhouette_score(filtered_features, filtered_labels)
    print("Silhouette Score (DBSCAN):", score)
else:
    print("Not enough clusters found for silhouette score.")
```

```python
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Purchase Amount (USD)'],
                y=df['Age'],
                hue=df['DBSCAN_Cluster'],
                palette='viridis')
plt.title("DBSCAN Clustering (Purchase Amount vs Age)")
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(features)
distances, indices = neighbors_fit.kneighbors(features)

# Sort distances and plot
distances = np.sort(distances[:, 4])  # 4th nearest neighbor
plt.plot(distances)
plt.title('K-distance Graph (for tuning eps)')
plt.xlabel('Points sorted by distance')
plt.ylabel('4th Nearest Neighbor Distance')
plt.grid()
plt.savefig('K_distance_graph')
plt.show()
```

#Unsupervised Learning Report: DBSCAN Clustering

# Objective
The goal was to segment mall customers into meaningful groups using an unsupervised learning technique. For this, we applied DBSCAN (Density-Based Spatial Clustering of Applications with Noise), which is effective for discovering clusters of arbitrary shape and detecting outliers in the data.

# Approach

We used customer attributes such as age, purchase behavior, and previous interactions to group similar customers. Unlike KMeans, DBSCAN does not require the number of clusters to be defined in advance. Instead, it uses two key parameters:

#eps (the radius of the neighborhood)

#min_samples (minimum number of points required to form a dense region)

These parameters were tuned manually to achieve the best clustering performance.

# Evaluation Metric: Silhouette Score
To evaluate the quality of the clusters, we used the Silhouette Score, which measures how well each data point fits within its assigned cluster compared to other clusters. The score ranges from:

+1 (well clustered)

0 (indistinct clusters)

-1 (incorrect clustering)

After experimentation, the best results were achieved with:

eps = 0.9

min_samples = 7

This produced a Silhouette Score of 0.556, which indicates a moderately strong cluster structure in the dataset.

# Applying Supervised Learning Techniques

```python
# Target
target = df['Subscription Status']

# Feature selection
features = df[['Age', 'Gender', 'Category', 'Purchase Amount (USD)',
               'Review Rating', 'Payment Method', 'Discount Applied',
               'Previous Purchases', 'Frequency of Purchases']]
```

```python
target.value_counts()
```

```python
features
```

```python
# normalize the dataset
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
scaled_features = ss.fit_transform(features)
```

```python
scaled_features.shape
```

#Train-Test Split

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features, target, test_size=0.2, random_state=42)
```

```python
y_train.shape, y_test.shape
```

#Build and Train Model

```python
# Build and Train the Model
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Create Pipeline
pipeline = Pipeline([
    ('classifier', RandomForestClassifier())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#Random Forest classifier Report
I trained a Random Forest classifier to predict whether a customer has an active subscription (Subscription Status).

Key performance metrics:

Precision (Subscribers): 64%

Recall (Subscribers): 86%

F1-score: 73%

The model is highly effective at identifying actual subscribers, which is valuable for marketing or retention strategies.
